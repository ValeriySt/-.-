```python
# Изучим зависимость цены квадратного метра от общей площади квартиры.
pivot_table_total_area = data.pivot_table(index = 'living_area_m2', values = 'price_per_meter', aggfunc = ['mean', 'count', 'median'])
pivot_table_total_area.columns = ['mean', 'count', 'median']
pivot_table_total_area.plot(y = 'median', style = 'o')

pivot_table_total_area.sort_values('median', ascending = False)

data['living_area_m2'].corr(data['price_per_meter'])
```

Корреляция положительная, соответственно чем больше квадратура, тем больше цена.
