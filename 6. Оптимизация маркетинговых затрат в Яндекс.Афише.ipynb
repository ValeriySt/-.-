2.1.4. Рассчитаем Retention Rate, применяя когортный анализ. Покажем изменения метрики во времени на графике. Найдем средний Retention Rate на второй месяц «жизни» когорт.

```python
visits.head(5)
```

```python
# Определим событие и период — на их основе сформируем когорту.
first_visit = visits.groupby(['uid'])['start_ts'].min().reset_index()
first_visit.columns = ['uid', 'first_visit']
first_visit.head(10)
```

```python
# Добавим столбец
visits = visits.merge(first_visit)
visits.head(10)
```

```python
visits['first_visit_month'] = visits['first_visit'].astype('datetime64[M]')
visits['visit_month'] = visits['start_ts'].astype('datetime64[M]')
```

```python
visits.head(10)
```

```python
visits['lifetime'] = (visits['visit_month'] - visits['first_visit_month']) / np.timedelta64(1, 'M')
visits['lifetime'] = visits['lifetime'].round().astype('int')
```

```python
# Чтобы найти Retention Rate, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждую следующую неделю.
# Найдём исходное количество пользователей в когорте.
cohort_visits = visits.groupby(['first_visit_month', 'lifetime']).agg({'uid': 'nunique'}).reset_index()
f_uid = cohort_visits[cohort_visits['lifetime'] == 0][['first_visit_month', 'uid']]
f_uid.columns = ['first_visit_month', 'f_uid']
f_uid.head(10)
```

```python
# Построим сводную таблицу 

cohort_visits = cohort_visits.merge(f_uid)
cohort_visits['retention'] = cohort_visits['uid'] / cohort_visits['f_uid']
visits_retention_pivot = cohort_visits.pivot_table(index = 'first_visit_month', 
                                                   columns = 'lifetime', values = 'retention', aggfunc = 'sum')

visits_retention_pivot.mean() * 100
```

```python
#находим время первого посещения
first_visits = visits.groupby('uid').agg({'start_ts': 'min'}).reset_index()
first_visits.columns = ['uid', 'first_visits_datetime']

#добавляем колонки с датой и месяцем
first_visits['first_visits_date'] = first_visits['first_visits_datetime'].dt.date
first_visits['first_visits_month'] = first_visits['first_visits_datetime'].astype('datetime64[M]')

#находим дату первой покупки
first_orders = orders.groupby('uid').agg({'purchase_ts': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_datetime']

#соединяем с первые показатели
orders_new = pd.merge(orders, first_orders, on='uid')

#доьавляем месяц покупки и первый месяц покупки
orders_new['order_month'] = orders_new['purchase_ts'].astype('datetime64[M]')
orders_new['first_order_month'] = orders_new['first_order_datetime'].astype('datetime64[M]')

#выделяем источник
users = visits.sort_values('start_ts').groupby('uid').first()
users = users[['source_id']]

#соединяем таблицы
orders_new = pd.merge(orders_new, users, left_on='uid', right_index=True)
```

```python
#переводим столбец в нужный формат
costs['costs_month'] = costs['costs_month'].astype('datetime64[M]')
```

```python
#группируем по источнику и месяцу первой покупки для уникальных uid
cohort_sizes_t = (
    orders_new.groupby(['source_id', 'first_order_month'])
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes_t.columns = ['source_id', 'first_order_month', 'n_buyers']

#группируем по источнику,месяцу покупки и месяцу первой покупки для суммарной выручки
cohorts_revenue_t = (
    orders_new.groupby(['source_id', 'first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
cohorts_revenue_t.columns = ['source_id', 'first_order_month', 'order_month', 'gp']

#соединяем
report_romi = pd.merge(cohort_sizes_t, cohorts_revenue_t, on=['source_id', 'first_order_month'])

#считаем 
report_romi['age'] = (
    report_romi['order_month'] - report_romi['first_order_month']
) / np.timedelta64(1, 'M')
report_romi['age'] = report_romi['age'].round().astype('int')
report_romi['ltv'] = report_romi['gp'] / report_romi['n_buyers']

#доьавляемрасходы
report_romi = pd.merge(report_romi,costs.groupby(['source_id', 'costs_month']).agg({'costs': 'sum'}).reset_index(),
               left_on=['source_id', 'first_order_month'], right_on=['source_id', 'costs_month'])
report_romi = report_romi.drop(['costs_month'], axis = 1) 

#еще считаем
report_romi['cac'] = report_romi['costs'] / report_romi['n_buyers']
report_romi['romi'] = report_romi['ltv'] / report_romi['cac']
display(report_romi.head())
```
